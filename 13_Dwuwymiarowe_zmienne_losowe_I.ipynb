`Niech $X$ oraz $Y$ będą zmiennymi losowymi określonymi niekoniecznie na tej samej przestrzeni probabilistycznej. Parę $(X, Y)$ zmiennych losowych $X, Y$ nazywamy **dwuwymiarową zmienną losową** lub **dwuwymiarowym wektorem losowym** a $X$ oraz $Y$ jej **współrzędnymi**.

**Dystrybuantą dwuwymiarowej zmiennej losowej $(X, Y)$** nazywamy funkcję $F$ zmiennych $x, y$, która dla każdej pary liczb rzeczywistych $(x, y) \in \mathbb{R^2}$ przyjmuje wartości równe prawdopodobieństwu zdarzenia polegającego na tym, że zmienna losowa $X$ przyjmuje wartość mniejszą od $x$ i zmienna losowa $Y$ przyjmuje wartość mniejszą od $y$:

$$F(x,y) = P(X<x, Y<y)\ \text{dla} \ (x, y) \in \mathbb{R^2}. \tag{1}$$

$F$ nazywamy także **dystrybuantą łącznej zmiennej losowej $(X, Y)$**.

Własności dystrybuanty:

a) $\forall x\in\mathbb{R} \lim\limits_{y \to -\infty} F(x, y) = 0, \ \ \ \ \ \forall y\in\mathbb{R} \lim\limits_{x \to -\infty} F(x, y) = 0,$

b) $\lim\limits_{x \to \infty \\ y \to \infty} F(x, y) = 1,$ $\tag{2}$

c) Dla dowolnych punktów: $(x_1, y_1)$, $(x_2, y_2)$ takich, że $x_1 \le x_2$ i $y_1 \le y_2$ zachodzi nierówność:

$F(x_2, y_2) - F(x_2, y_1) - F(x_1, y_2) + F(x_1, y_1) \ge 0$.

d) Dystrybuanta jest funkcją niemalejącą i co najmniej lewostronnie ciągłą względem każdego z argumentów $x$ bądź $y$.

Dwuwymiarową zmienną losową $(X, Y)$, która przyjmuje skończoną, bądź przeliczalną liczbę wartości $(x_i, y_k)$, każdą odpowiednio z prawdopodobieństwem:

$P(X = x_i, Y = y_k) = p_{ik} \ \ \text{dla} \ \ i, k \in\mathbb{N}, \tag{3}$

przy czym $\sum_\limits{i}\sum_\limits{k}p_{ik} = 1,$ nazywamy **dwuwymiarową zmienną losową skokową (dyskretną)**.

Oznaczmy:

$p_{i.} = \sum\limits_{k}p_{ik} \ \ \text{dla} \ \ i \in\mathbb{N}, \tag{4}$
$p_{.k} = \sum\limits_{i}p_{ik} \ \ \text{dla} \ \ k \in\mathbb{N}. \tag{5}$

$p_{i.} = P(X=x_i, Y=y_1) + P(X=x_i, Y=y_2)+... \ $ jest prawdopodobieństwem tego, że zmienna losowa $X$ przyjmuje wartość równą $x_i$, bez względu na to, którą z wartości: $y_1, y_2,...$ przyjmuje zmienna losowa $Y$, oraz, że $\sum\limits_i p_{i.} = 1,$ a więc funkcja:

$P(X=x_i)=p_{i.}, \ \ i\in\mathbb{N} \tag{6}$

wyznacza rozkład prawdopodobieństwa zmiennej losowej $X$, nazywamy **rozkładem brzegowym** zmiennej $X$ w rozkładzie dwuwymiarowej zmiennej losowej $(X, Y)$. Analogicznie rozkład brzegowy zmiennej losowej $Y$ określamy wzorem:

$P(Y=y_k)=p_{.k}, \ \ k\in\mathbb{N}. \tag{7}$

Jeśli dwuwymiarowa zmienna losowa $(X, Y)$ przyjmuje skończoną liczbę wartości, to wygodnie jest umieścić wartość funkcji prawdopodobieństwa w tabelce dwudzielczej:

|||||||
|---|---|---|---|---|---|
||||$x_i$|||
|$y_k$|$x_1$|$x_2$|$...$|$x_m$|$p_{.k}$|
|$y_1$|$p_{11}$|$p_{21}$|$...$|$p_{m1}$|$p_{.1}$|
|$y_2$|$p_{12}$|$p_{22}$|$...$|$p_{m2}$|$p_{.2}$|
|$...$|$...$|$...$|$...$|$...$|$...$|
|$y_s$|$p_{1s}$|$p_{2s}$|$...$|$p_{ms}$|$p_{.s}$|
|$p_{i.}$|$p_{1.}$|$p_{2.}$|$...$|$p_{m.}$|$1$|


Oznaczmy przez $F_1$ i $F_2$ dystrybuanty rozkładów brzegowych zmiennych losowych $X$ i $Y$ odpowiednio. Jeśli $(X, Y)$ jest dwuwymiarową zmienną losową skokową, to:

$F_1(x) = \sum\limits_{x_i<x}p_{i.} \ \ \ \text{dla} \ \ \ x\in\mathbb{R}, \tag{8}$

$F_2(y) = \sum\limits_{y_k<y}p_{.k} \ \ \ \text{dla} \ \ \ y\in\mathbb{R}, \tag{9}$

**Rozkład warunkowy** zmiennej losowej $X$, pod warunkiem, że $Y$ przyjmuje ustaloną wartość równą $y_k$ określamy wzorem:

$P(X=x_i|Y=y_k) = \frac{p_{ik}}{p_{.k}}, \ \ \ k\in\mathbb{N}. \tag{10}$

Analogicznie można określić rozkład prawdopodobieństwa zmiennej losowej $Y$ przy warunku, że $X$ przyjmuje wartość $x_i$: 

$P(Y=y_k|X=x_i) = \frac{p_{ik}}{p_{i.}}, \ \ \ k\in\mathbb{N}. \tag{11}$

Dystrybuanty rozkładów warunkowych oznaczamy odpowiednio przez: $F(x|y_k), \ F(y|x_i) i wyznaczamy ze wzorów:$

$F(x|y_k) = P(X<x|Y=y_k) = \sum\limits_{x_i<x} \frac{p_{ik}}{p_{.k}}, \tag{12}$

$F(y|x_i) = P(Y=y|X=x_i) = \sum\limits_{y_k<y} \frac{p_{ik}}{p_{i.}}. \tag{13}$

**Dwuwymiarową zmienną losową $(X, Y)$** nazywamy **typu ciągłego**, jeśli istnieje nieujemna funkcja $f$ taka, że dystrybuanta tej zmiennej losowej da się przedstawić jako całka:

$F(x, y) = \int\limits_{-\infty}^x \int\limits_{-\infty}^y f(u, v) dvdu, \ \ \ \text{dla} \ \ \ (x, y) \in\mathbb{R^2}, \tag{14}$

funkcję $f$ nazywamy **gęstością rozkładu prawdopodobieństwa**.

Własności dwuwymiarowej zmiennej losowej typu ciągłego:

1) $\int\limits_{-\infty}^\infty \int\limits_{-\infty}^\infty f(x, y) dydx = 1, \tag{15}$

2) w punktach $(x, y)$ ciągłości $f$ mamy

$\frac{\partial^2F(x, y)}{\partial x \partial y} = f(x, y), \tag{16}$

3) dla obszaru regularnego $B \subset \mathbb{R^2}$

$P[(X, Y) \in B] = \iint\limits_{B} f(x,y)dxdy, \tag{17}$

w szczególności gdy $B$ jest prostokątem, tzn. $B = \{(x, y): a \le x \le b \land c \le y \le d\}$, wtedy

$P(a \le X \le b \land c \le Y \le d) = \int\limits_a^b \int\limits_c^d f(x, y) dydx.$

**Rozkład brzegowy** zmiennej losowej $X$ w rozkładzie dwuwymiarowej zmiennej losowej $(X, Y)$ określa wzór:

$f_1(x) = \int\limits_{-\infty}^\infty f(x, y) dy. \tag{18}$

**Dystrybuantę rozkładu brzegowego** zmiennej losowej $X$ (ciągłej) w rozkładzie dwuwymiarowym zmiennej losowej $(X, Y)$ wyznacza wzór:

$F_1(x) = \int\limits_{-\infty}^x \int\limits_{-\infty}^\infty f(u, y) dy du = \int\limits_{-\infty}^x f_1(u)du \ \ \ \text{dla} \ \ \ x\in\mathbb{R}. \tag{19}$

Analogicznie rozkład brzegowy zmiennej losowej $Y$ w rozkładzie dwuwymiarowej zmiennej losowej $(X, Y)$ określa wzór:

$f_2(y) = \int\limits_{-\infty}^\infty f(x, y) dx. \tag{20}$

Dystrybuantę rozkładu brzegowego zmiennej losowej $Y$ (ciągłej) w rozkładzie dwuwymiarowym zmiennej losowej $(X, Y)$ wyznacza wzór:

$F_2(y) = \int\limits_{-\infty}^y \int\limits_{-\infty}^\infty f(x, v) dx dv = \int\limits_{-\infty}^y f_2(v)dv \ \ \ \text{dla} \ \ \ y\in\mathbb{R}. \tag{21}$

Związek między dystrybuantami rozkładów brzegowych, a dystrybuantą $F$ dwuwymiarowej zmiennej losowej $(X, Y)$, a mianowicie:

$\begin{aligned}
F_1(x) = \lim\limits_{y \to \infty} F(x, y), \\
F_2(y) = \lim\limits_{x \to \infty} F(x, y).
\end{aligned} \tag{22}$

**Gęstość warunkową** zmiennej losowej $X$ przy warunku, że zmienna $Y$ przyjęła wartość równą $y$ określa wzór:

$f(x|y) = \frac{f(x, y)}{f_2(y)} \ \ \ \text{dla} \ \ \ x\in\mathbb{R} \land c < y < d, \tag{23}$

przy założeniu, że gęstość brzegowa $f_2$ jest skoncentrowana na przedziale $(c, d)$.

Dystrybuantę rozkładu warunkowego wyznaczamy według wzoru:

$F(x|y) = \int\limits_{-\infty}^x f(u|y)du = \frac{1}{f_2(y)} \int\limits_{-\infty}^x f(u, y)du \ \ \ \text{dla} \ \ \ x\in\mathbb{R} \land c < y < d. \tag{24}$


Analogicznie Gęstość warunkową zmiennej losowej $Y$ przy warunku, że zmienna $X$ przyjęła wartość równą $x$ określa wzór:

$f(y|x) = \frac{f(x, y)}{f_1(x)} \ \ \ \text{dla} \ \ \ y\in\mathbb{R} \land a < x < b, \tag{25}$

przy założeniu, że gęstość brzegowa $f_1$ jest skoncentrowana na przedziale $(a, b)$.

Dystrybuantę tego rozkładu warunkowego wyznaczamy według wzoru:

$F(y|x) = \int\limits_{-\infty}^y f(v|x)dv = \frac{1}{f_1(x)} \int\limits_{-\infty}^y f(x, v)dv \ \ \ \text{dla} \ \ \ y\in\mathbb{R} \land a < x < b. \tag{26}$

###Niezależność zmiennych losowych
1) Warunkiem koniecznym i wystarczającym na to, by $X$ i $Y$ były niezależnymi zmiennymi losowymi jest, by dla każdego $(x,y) \in \mathbb{R^2}$ dystrybuanta $F$ dwuwymiarowej zmiennej losowej $(X,Y)$ była iloczynem dystrybuant rozkładów brzegowych $F_1$ i $F_2$:

$F(x,y) = F_1(x)F_2(y). \tag{27}$

2) Warunkiem koniecznym i wystarczającym niezależności zmiennych losowych ciągłych $X, Y$ o gęstościach ospowiednio równych $f_1, f_2$ jest zachodzenie równości: 

$f(x,y) = f_1(x)f_2(y) \ \ \ \text{dla} \ \ \ x, y \in\mathbb{R}, \tag{28}$

gdzie $f$ jest gęstością dwuwymiarowej zmiennej losowej $(X,Y)$.


Zmienne losowe $X$ i $Y$ typu skokowego są niezależne wtedy i tylko wtedy, gdy

$P(X=x_i, Y=y_k) = P(X=x_i)P(Y=y_k) \ \ \ \text{dla} \ \ \ i,k\in\mathbb{N}, \tag{29}$

czyli:

$$p_{ik} = p_{i.}p_{.k} \ \ \ \text{dla} \ \ \ i,k\in\mathbb{N}.$$

3) $X$ i $Y$ są niezależnymi zmiennymi losowymi wtedy i tylko wtedy, gdy rozkłady warunkowe są równe odpowiednim rozkładom brzegowym, co można zapisać np. tak:

$$F(y|x) = F_2(y), \ \ \ F(x|y)=F_1(x).$$
